## Frequency count for categorical data
    
how can you diagnose data issues when you have categorical data

By using the `.value_counts()` method, which returns the frequency counts for each unique value in a column!


This method also has an optional parameter called dropna which is True by default. What this means is if you have missing data in a column, it will not give a frequency count of them. You want to set the dropna column to False so if there are missing values in a column, it will give you the frequency counts.

```python
df.Borough.value_counts()
```

```python
df.Borough.value_counts(dropna=False)
```

```python
df.State.value_counts(dropna= False)
```

```python
df['Site Fill'].value_counts(dropna = False)
```

Fantastic work! Notice how not all values in the 'State' column are NY. This is an interesting find, as this data is supposed to consist of applications filed in NYC. Curiously, all the 'Borough' values are correct. A good start as to why this may be the case would be to find and look at the codebook for this dataset. Also, for the 'Site Fill' column, you may or may not need to recode the NOT APPLICABLE values to NaN in your final analysis.

### Reshaping data using `melt`
    
    
Melting data is the process of turning columns of your data into rows of data. 

There are two parameters you should be aware of: id_vars and value_vars.  
The `id_vars` represent the columns of the data you **do not** want to melt (i.e., keep it in its current shape),  
while the `value_vars` represent the columns you **do** wish to melt into rows.   
By default, if no value_vars are provided, all columns not set in the id_vars will be melted. This could save a bit of typing, depending on the number of columns that need to be melted.



```python
airquality  = pd.read_csv('./data/airquality.csv')
```

```python
airquality.head()
```

```python
airquality_melt = pd.melt(frame = airquality, id_vars = ['Month', 'Day'] )
airquality_melt.head()
```

<a id = 'cmelt'> <a>
### Customizing melted data
    
    
When melting DataFrames, it would be better to have column names more meaningful than variable and value (the default names used by pd.melt()).

```python
airquality_melt = pd.melt(frame = airquality, id_vars = ['Month', 'Day'], var_name= 'measurement', value_name='reading' )
airquality_melt.head()
```

<a id= 'pivot'> <a>
## Pivoting Data: Un-melting Data
    
Turning unique values into separate colums


We can see how min and max reported in separate rows, causing dates to be **DUPLICATED**


` .pivot_table()` has an `index` parameter which we can use to specify the columns that you don't want pivoted:   
It is similar to the `id_vars` parameter of `pd.melt()`. Two other parameters that you have to specify are `columns` (the name of the column you want to pivot), and `values` (the values to be used when the column is pivoted).


```python
airquality_melt.head()
```

```python
# Pivot airquality_melt: airquality_pivot
airquality_pivot = airquality_melt.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading')
```

<a id = 'reset'> <a>
### Reseting Index of data

```python
airquality_pivot.head()
```

```python
airquality_pivot.index
```

```python
# Reset the index of airquality_pivot: airquality_pivot_reset
airquality_pivot_reset = airquality_pivot.reset_index()
```

```python
airquality_pivot_reset.head()
```

```python
print(airquality_pivot_reset.index)
```

You'll see that by using .pivot_table() and the aggfunc parameter, you can not only reshape your data, but also remove duplicates. Finally, you can then flatten the columns of the pivoted DataFrame using .reset_index().



<a id = 'dupl'> <a>
### Pivoting Duplicate Values

```python
# Creating Observation with Duplicates  values

print(airquality_melt.shape)
airquality_dup = pd.concat([airquality_melt, airquality_melt])

print()
print(airquality_dup.shape)

```

We can also use pivot tables to deal with duplicate values by providing an aggregation function through the `aggfunc` parameter. 

```python
# Pivot table the airquality_dup: airquality_pivot
airquality_pivot = airquality_dup.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading', aggfunc=np.mean)

# Print the head of airquality_pivot before reset_index
print(airquality_pivot.head())
```

```python
# Reset the index of airquality_pivot
airquality_pivot = airquality_pivot.reset_index()

# Print the head of airquality_pivot
print(airquality_pivot.head())
```

```python
# Print the head of airquality
print(airquality.head())
```

### Splitting a column with `.str`

```python
tb = pd.read_csv('./data/tb.csv')
```

```python
tb.head()
```

In this exercise, We're going to tidy the `'m014'` column, which represents males aged 0-14 years of age.   
In order to parse this value, you need to extract the first letter into a new column for gender, and the rest into a column for age_group. Here, since you can parse values by position, you can take advantage of pandas' vectorized string slicing by using the str attribute of columns of type object.



```python
tb_melt = pd.melt(frame = tb, id_vars = ['country', 'year'])
tb_melt.head()
```

```python
# Create the 'gender' column
tb_melt['gender'] = tb_melt.variable.str[0]
tb_melt.head()
```

```python
# Create the 'age_group' column
tb_melt['age_group'] = tb_melt.variable.str[1:]
tb_melt.head()
```

<p id = 'split'>  <p>
### Splitting a column with `.split()` and `.get()`
    
    
Another common way multiple variables are stored in columns is with a delimiter. 

```python
ebola = pd.read_csv('./data/ebola.csv')
```

```python
ebola.head()
```

```python
ebola.columns
```

```python
ebola_melt = pd.melt(ebola, id_vars=['Date', 'Day'], var_name='type_country', value_name='counts')
ebola_melt.head()
```

```python
# Create the 'str_split' column
ebola_melt['str_split'] = ebola_melt['type_country'].str.split('_')
ebola_melt.head()
```

```python
# Create the 'type' column
ebola_melt['type'] = ebola_melt['str_split'].str.get(0)
ebola_melt.head()
```

```python
# Create the 'country' column
ebola_melt['country'] = ebola_melt['str_split'].str.get(1)
ebola_melt.head()
```
