### Optimal parameters
Parameter values that bring the model in closest agreement with data

```python
from data.nohitter_times import nohitter_times
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
def pearson_r(x, y):
    return np.corrcoef(x, y)[0,1]
```

```python
def Ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y
```

```python
print(nohitter_times[:5])
print(nohitter_times.size)
```

<p id ='Hodwgn'><p>
### How often do we get no-hitters?
   
If you assume that no-hitters are described as a Poisson process, then the time between no-hitters is Exponentially distributed. As you have seen, the Exponential distribution has a single parameter, which we will call τ
, the typical interval time. The value of the parameter τ
 that makes the exponential distribution best match the data is the mean interval time (where time is in units of number of games) between no-hitters.



```python
np.random.seed(42)
```

```python
tau = np.mean(nohitter_times)
tau
```

```python
inter_nohitter_time = np.random.exponential(tau, size = 10000)
```

```python
min(inter_nohitter_time)
```

```python

# Plot the PDF and label axes
_ = plt.hist(inter_nohitter_time,
             bins = 50, density = True, histtype = 'step')
_ = plt.xlabel('Games between no-hitters')
_ = plt.ylabel('PDF')
```

<p id ='Dtdfos'><p>
### Do the data follow our story?
You have modeled no-hitters using an Exponential distribution. Create an ECDF of the real data. Overlay the theoretical CDF with the ECDF from the data. This helps you to verify that the Exponential distribution describes the observed data.



```python
# Create an ECDF from real data: x, y
x, y = Ecdf(nohitter_times)
```

```python
# Create a CDF from theoretical samples: x_theor, y_theor
x_theor, y_theor = Ecdf(inter_nohitter_time)
```

```python
plt.plot(x_theor, y_theor)
plt.plot(x, y, marker = '.', linestyle = 'none')

# Margins and axis labels
plt.margins(0.02)
plt.xlabel('Games between no-hitters')
plt.ylabel('CDF')

# Show the plot
plt.show()
```

It looks like no-hitters in the modern era of Major League Baseball are Exponentially distributed. Based on the story of the Exponential distribution, this suggests that they are a random process; when a no-hitter will happen is independent of when the last no-hitter was.



<p id ='Hitpo'><p>
### How is this parameter optimal?
Now sample out of an exponential distribution with τ
 being twice as large as the optimal τ
. Do it again for τ
 half as large. Make CDFs of these samples and overlay them with your data. You can see that they do not reproduce the data as well. Thus, the τ
 you computed from the mean inter-no-hitter times is optimal in that it best reproduces the data.



```python
# Take samples with half tau: samples_half
samples_half = np.random.exponential(tau/2, size = 10000)

# Take samples with double tau: samples_double
samples_double = np.random.exponential(2*tau, size = 10000)

# Generate CDFs from these samples
x_half, y_half = ecdf(samples_half)
x_double, y_double = ecdf(samples_double)


# Plot the theoretical CDFs
plt.plot(x_theor, y_theor)
plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('Games between no-hitters')
plt.ylabel('CDF')



# Plot these CDFs as lines
_ = plt.plot(x_half, y_half)
_ = plt.plot(x_double, y_double)

# Show the plot
plt.show()
```

<p id ='Lr'><p>
### Linear regression
We will assume that fertility is a linear function of the female illiteracy rate. That is, `f=ai+b`
, where a
 is the slope and b
 is the intercept.    
* We can think of the intercept as the minimal fertility rate, probably somewhere between one and two.
* The slope tells us how the fertility rate varies with illiteracy. We can find the best fit line using `np.polyfit()`.


* **Least Squares**: The process of finding the parameters for which the sum of squares of the residuals is minimum 


```python
a, b = np.polyfit(df.illiteracy, df.fertility, 1)
```

```python
# Print the results to the screen
print('slope =', a, 'children per woman / percent illiterate')
print('intercept =', b, 'children per woman')
```

```python
x = np.array((0, 100))
y = a * x + b

```

```python
# Plot the illiteracy rate versus fertility
_ = plt.plot(df.illiteracy, df.fertility, marker='.', linestyle='none')
plt.margins(0.02)
_ = plt.xlabel('percent illiterate')
_ = plt.ylabel('fertility')

# Add regression line to your plot
_ = plt.plot(x, y)
```

<p id ='Hiio'><p>
### How is it optimal?
    
The function `np.polyfit()` that you used to get your regression parameters finds the optimal slope and intercept. It is optimizing the sum of the squares of the residuals, also known as RSS (for residual sum of squares). In this exercise, you will plot the function that is being optimized, the RSS, versus the slope parameter a. To do this, fix the intercept to be what you found in the optimization. Then, plot the RSS vs. the slope. Where is it minimal?



```python
# Specify slopes to consider
a_vals = np.linspace(0, 0.1, 200)
```

```python
# Initialize sum of square of residuals: rss
rss = np.empty_like(a_vals)

```

```python
np.sum((df.fertility- a*df.illiteracy-b)**2)
```

```python
for i, a in enumerate(a_vals):
    rss[i] = np.sum((df.fertility- a*df.illiteracy-b)**2)
```

```python
plt.plot(a_vals, rss, '-')
plt.xlabel('slope (children per woman / percent illiterate)')
plt.ylabel('sum of square of residuals')

```

### Generating bootstrap replicates
Michaleson was not interested in what the measured speed of light was in the specific 100 measurements conducted in the summer of 1879.
    
Statistically speaking he wants to know what speed of light he would observe, if he did the experiments over and over again an infinite number of times.



* The idea is that we resample the data we have and recompute the summary statistic of interest, say mean

*  **Bootstrap Replicate**: A statistic computed from a resampled array.

### Bootstrapping: The use of resampled data to perform statistical inference

### Visualizing bootstrap samples
We will generate bootstrap samples from the set of annual rainfall data measured at the Sheffield Weather Station in the UK from 1883 to 2015. By graphically displaying the bootstrap samples with an ECDF, you can get a feel for how bootstrap sampling allows probabilistic descriptions of data.

```python
from data.rainfall import rainfall
```

```python
bs_sample = np.random.choice(rainfall, size = len(rainfall))
```

```python


for i in range(50):
    # Generate bootstrap sample: bs_sample
    bs_sample = np.random.choice(rainfall, size = len(rainfall))

    # Compute and plot ECDF from bootstrap sample
    x, y = ecdf(bs_sample)
    _ = plt.plot(x, y, marker='.', linestyle='none',
                 color='gray', alpha=0.1)

# Compute and plot ECDF from original data
x, y = ecdf(rainfall)
_ = plt.plot(x, y , marker='.', color = 'red')

# Make margins and label axes
plt.margins(0.02)
_ = plt.xlabel('yearly rainfall (mm)')
_ = plt.ylabel('ECDF')

# Show the plot
plt.show()
```

<p id ='Bci'><p>
### Bootstrap confidence intervals
If we repeated measurements over and over again, p% of observed values would lie within the p% confidence interval.  

```python
def bootstrap_replicate_1d(data, func):
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)
```

```python
bootstrap_replicate_1d(rainfall, np.mean )
```

<p id ='Gmbr'><p>
### Generating many bootstrap replicates

```python
def draw_bs_reps(data, func, size = 1):
    """Draw bootstrap replicates"""
    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)
      # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)
    return bs_replicates
```

```python
# let's see how above above func works
arr = draw_bs_reps(rainfall, np.mean, 1000)
print(arr.shape)
# array arr will contain 10000 bootstrap replicate, i.e. single value stats (mean) for 10000 samples
print(arr[:10])
```

### Bootstrap replicates of the mean and the SEM
SEM : Standard Error of Mean  
We will compute **a bootstrap estimate of the probability density function ** of the mean annual rainfall at the Sheffield Weather Station.        
    We are estimating the mean annual rainfall we would get if the Sheffield Weather Station could repeat all of the measurements from 1883 to 2015 over and over again. 

```python
bs_replicates = draw_bs_reps(rainfall, np.mean, 10000)
```
