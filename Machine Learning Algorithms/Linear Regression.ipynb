# Linear Regression 
#### Make neccessary imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Reading Fruits Data

```python
plt.figure()
plt.title('Sample regression problem with one input variable')
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, bias =150.0, noise =30, random_state=0)
plt.scatter(X, y, marker= 'o', s= 50)
plt.show()
```

### Splitting the data in train and test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

### Fitting the Model on training set

```python
linreg = LinearRegression().fit(X_train, y_train)
```

### Visualizing the model performance

```python
print('linear model coeffient (w):{}'. format(linreg.coef_))
print('linear model intercept (w):{}'. format(linreg.intercept_))
```

```python
plt.scatter(X, y, marker= 'o', alpha = 0.8)
plt.plot(X, linreg.coef_*X+linreg.intercept_, 'r-')
plt.title('Least-squares linear regression')
plt.xlabel('Feature value (x)')
plt.ylabel('Target value (y)')
plt.show()
```

The coefficient of determination, denoted R2 or r2 and pronounced "R squared", is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
How well predicted line fit to the model.

### Evaluating the performance of the model on the test set

```python
print(' R square Score on test set: ',linreg.score(X_test, y_test))
```

#### Ridge regression with feature normalization

### Ridge Regression with regularization parameter


```python
print('Ridge regression: effect of alpha regularization parameter\n')

for this_alpha in [0, 1, 10, 20, 50, 100, 1000]:
    linridge = Ridge(alpha = this_alpha).fit(X_train_scaled, y_train)
    r2_train = linridge.score(X_train_scaled, y_train)
    r2_test = linridge.score(X_test_scaled, y_test)
    num_coeff_bigger = np.sum(abs(linridge.coef_)>1.0)
    print('Alpha = {:.2f}\nnum abs(coeff) >1.0: {} ,\
    r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_alpha, num_coeff_bigger, r2_train, r2_test))
```

### Addition of many polynomial features often leads to overfitting, so we often use polynomial features in combination with regression that has a regularization penalty, like ridge regression.

```python
linreg = Ridge().fit(X_train, y_train)

print('(poly deg 2 + ridge) linear model coeff (w):\n{}'
     .format(linreg.coef_))
print('(poly deg 2 + ridge) linear model intercept (b): {:.3f}'
     .format(linreg.intercept_))
print('(poly deg 2 + ridge) R-squared score (training): {:.3f}'
     .format(linreg.score(X_train, y_train)))
print('(poly deg 2 + ridge) R-squared score (test): {:.3f}'
     .format(linreg.score(X_test, y_test)))
```
